```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset (replace with your own dataset path if necessary)
df = sns.load_dataset('tips')

# Data Storytelling
print("Dataset Overview")
print(df.head())

# 1. Descriptive Statistics
print("\nDescriptive Statistics")
print(df.describe())

# 2. Distribution of Total Bill
plt.figure(figsize=(10, 6))
sns.histplot(df['total_bill'], kde=True, color='blue')
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()

# 3. Tips by Day
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='tip', data=df, palette='Set3')
plt.title('Tips by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Tip Amount')
plt.show()

# 4. Relationship between Total Bill and Tip
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', hue='day', data=df, palette='Set1')
plt.title('Relationship between Total Bill and Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()

# 5. Average Tip by Sex and Smoker Status
plt.figure(figsize=(10, 6))
sns.barplot(x='sex', y='tip', hue='smoker', data=df, palette='muted')
plt.title('Average Tip by Sex and Smoker Status')
plt.xlabel('Sex')
plt.ylabel('Average Tip')
plt.show()

# 6. Correlation Heatmap
correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
